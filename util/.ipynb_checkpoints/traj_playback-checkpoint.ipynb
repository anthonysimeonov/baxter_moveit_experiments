```python
def collision_check(waypoint):
    if not isinstance(waypoint, (list,)): waypoint.tolist() 
    filler_robot_state = list(robot_state.joint_state.position)
#     filler_robot_state[10:17] = waypoint
    filler_robot_state[1:8] = waypoint # left
    rs_man.joint_state.position = tuple(filler_robot_state)
    return sv.getStateValidity(rs_man, group_name='right_arm')

def check_intersect(wp1, wp2):
    t = np.linspace(0, 1, 10)
    interp = [np.asarray(wp2)*i + (1-i)*np.asarray(wp1) for i in t]
    interp_collision = [collision_check(x) for x in interp]
    return all(interp_collision)

def traj_rewiring(path):
    for i in range(len(path) - 1):
        for j in range(len(path) - 1, i + 1, -1):
            if check_intersect(path[i], path[j]):
                pc = []
                for k in range(0, i + 1):
                    pc.append(path[k])
                for k in range(j, len(path)):
                    pc.append(path[k])
                return traj_rewiring(pc)
    return path

def path_interpolator(path, padding):
    t = np.linspace(0, 1, padding)
    interpolated_path = []
    for i in range(len(path) - 1):
        interpolated_path += [np.asarray(path[i+1])*j + (1-j)*np.asarray(path[i]) for j in t]
    return np.asarray(interpolated_path)

def traj_playback(path, limb_name):
    '''path is a list'''
    if not isinstance(path, (list,)): path.tolist()
    for i in range(len(path)):
        traj = {}
        for key, val in zip(group_plan_names[limb_name], path[i]):
            traj[key] = val
        limb_set[limb_name].set_joint_positions(traj)
        if i==0: rospy.sleep(5)
        rospy.sleep(0.05)

import pickle 

limb_set = {}
limb_set['right'] = baxter_interface.Limb('right')
# limb_set['left'] = baxter_interface.Limb('left')
        
# search_path = '/home/arclab-testchamber/Google Drive/UCSD stuff/ARCLab/Code/Python experiments/'
# # path_left = np.load(search_path+'cleaned_rollout_right_2.npy').tolist()
# path_left = np.load(search_path+'baxter_rollout_left.npy').tolist()
# path_right = np.load(search_path+'baxter_rollout_1.npy').tolist()
with open ("pathsTest_1.pkl", "rb") as r_paths_f:
    path_right = pickle.load(r_paths_f)[0]

# rewired_path_left = path_left
rewired_path_right = path_right

# rewired_path = traj_rewiring(path)
# rewired_path_left = traj_rewiring(path_left)

# rewired_path_right = traj_rewiring(path_right)

# print(len(rewired_path_left))
print(len(rewired_path_right))
```

```python
from baxter_pykdl import baxter_kinematics
# limb_name = 'left'
limb_name = 'right'

def cartesian_cost(x):
    cost = sum([np.linalg.norm(np.asarray(x[i]) - np.asarray(x[i+1])) for i in range(len(x)-1)])
    return cost

def joints2carteul(joints):
    traj = {}
    x = []
    for i in range(len(joints)):
        
        for key, val in zip(limb_set[limb_name].joint_names(), joints[i]):
            traj[key] = val
        carteul = np.zeros(6)
        carteul[0:3] = kin.forward_position_kinematics(traj)[0:3]
        x.append(carteul)
    return x

current_joints = list(rewired_path_right)
print(len(current_joints[2]))
kin = baxter_kinematics(limb_name)
alpha = 0.5

def inv_jacobian(jac_joints):
#     print(jac_joints)
    traj = {}
    for key, val in zip(limb_set[limb_name].joint_names(), jac_joints):
        traj[key] = val
#     print(traj)
    inv_jac = kin.jacobian_pseudo_inverse(mode='svd', joint_values=traj)
    return inv_jac

iterations = 0
cost = []
while True:
    for i in range(1, len(current_joints)-1):
        current_cartesian = joints2carteul(current_joints)
#         print(current_cartesian[i-1] + current_cartesian[i+1] - 2*current_cartesian[i])
        current_cost = cartesian_cost(current_cartesian)
        cost.append(current_cost)
        delta_x = (np.asarray(current_cartesian[i-1]) + np.asarray(current_cartesian[i+1])
                   - 2*np.asarray(current_cartesian[i]))/2
        delta_x = np.expand_dims(delta_x, axis=1)
#         print(delta_x.shape)
#         print(current_joints[i]))
        delta_q = alpha*np.matmul(inv_jacobian(current_joints[i]), delta_x)
        delta_q = [q[0] for q in delta_q.tolist()]
        q_hat = [m+ n for m, n in zip(current_joints[i], delta_q)]
#         print('q_hat', type(q_hat))
#         print(collision_flag(q_hat))
#         print(q_hat)
        if collision_check(q_hat):
            current_joints[i] = list(q_hat)
    iterations += 1
    print(iterations)
    step = np.linalg.norm(current_cost - cartesian_cost(joints2carteul(current_joints)))
    print(step)
    if iterations > 200:
        print('break?')
        break
plt.figure()
plt.plot(np.asarray(cost))
```

```python
import math
from scipy import interpolate

def find_segments(path):
    return [np.linalg.norm(np.asarray(path[i])-np.asarray(path[i+1])) 
                       for i in range(len(path)-1)]

def scaled_lengths(path):
    segment_lengths = find_segments(path)
    return [0]+[segment_lengths[i]/sum(segment_lengths) for i in range(len(segment_lengths)-1)]+[1]

def sigmoid(x):
    return 1 / (1 + math.exp(-10*(x-0.5)))

tx = rewired_path_right
t = np.linspace(-1, 1, 200)
tx = scaled_lengths(rewired_path_right)
txnew = [sigmoid(i) for i in t]
print(np.asarray(tx).shape)
print(np.asarray(rewired_path_right).shape)
plt.figure()
plt.plot(txnew)

# print(txnew)
f = interpolate.interp1d(np.asarray(tx), np.asarray(rewired_path_right), axis=0)
rescaled_path = f(txnew)
print(rescaled_path.shape)
plt.plot(find_segments(rescaled_path))
```
