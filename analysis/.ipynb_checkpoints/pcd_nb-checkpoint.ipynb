```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pickle
import pypcd

%matplotlib notebook
```

```python
pc = pypcd.PointCloud.from_path('../data/full_dataset_sample/trainEnv_0_185647000.pcd')
```

```python
pc_full = pc.pc_data.copy()
print(len(pc_full['x']))
sub = 10
print(len(pc_full['x'][::sub]))
```

```python
pc_sub_x = pc_full['x'][::sub]
pc_sub_y = pc_full['y'][::sub]
pc_sub_z = pc_full['z'][::sub]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# ax = Axes3D(fig)
ax.scatter(pc.pc_data['x'], pc.pc_data['y'], pc.pc_data['z'], color='b')
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')
# ax = Axes3D(fig)
ax1.scatter(pc_sub_x, pc_sub_y, pc_sub_z, color='b')
```

```python
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
